```python
#What factors are associated with higher attrition rates?
sns.set(style="whitegrid")
factors = ['Age', 'MonthlyIncome', 'DistanceFromHome', 'JobSatisfaction', 'JobInvolvement']

for factor in factors:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Attrition', y=factor, data=df)
    plt.title(f'Attrition vs {factor}')
    plt.show()
```

```python
#Q: What is the Attrition Rate Department wise? 
sns.countplot(x='Attrition', hue='Department', data=df,palette='bright')
plt.title('Attrition vs Department')
plt.show()
```

```python
#Q: Attrition Rate of different Job Roles and descriptions?
sns.countplot(x='Attrition', hue='JobRole', data=df)
plt.title('Attrition vs Job Role')
plt.show()
```

```python
#Q: Employees who work more are more likely to leave the company or not?
sns.countplot(x='Attrition', hue='OverTime', data=df,palette='dark')
plt.title('Attrition vs OverTime')
plt.show()
```

```python
#Q: Does Environment Satisfaction at job has any impact on Attrition or not?
sns.countplot(x='Attrition', hue='EnvironmentSatisfaction', data=df,palette='tab10')
plt.title('Attrition vs EnvironmentSatisfaction')
plt.show()
```

# Employee Training and Development

```python
#How many training times did employees have last year?
df['TrainingTimesLastYear'].sum()
```

Employees were trained 4115 times in a year.

```python
# Is there any co-relation b/w Employee Training and Performace of Employees at job?
correlation_matrix = df[['TrainingTimesLastYear', 'PerformanceRating']].corr()

plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', linewidths=.5)
plt.title('Correlation between Training and Performance')
plt.show()
```

```python
#Are employees with more training more likely to stay?
plt.figure(figsize=(6, 4))
sns.countplot(x='TrainingTimesLastYear', hue='Attrition', data=df)
plt.title('Attrition vs. Training Times Last Year')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(df.corr(), annot= True,cmap='mako')
plt.title('Correlation')
plt.show()
```

Here we can see their is a high correlation between joblevel,monthly income,total working years and job level also with monthly incom & total working.
