```python
# Rename a single column
df.rename(columns={'Aircraft Type': 'Aircraft_Type'}, inplace=True)
df.rename(columns={'Seat Comfort': 'Seat_Comfort'}, inplace=True)
df.rename(columns={'Date Flown': 'Date_Flown'}, inplace=True)
df.rename(columns={'Cabin Staff Service': 'Cabin_Staff_Service'}, inplace=True)
df.rename(columns={'Ground Service': 'Ground_Service'}, inplace=True)
df.rename(columns={'Food & Beverages': 'Food_Beverages'}, inplace=True)
df.rename(columns={'Wifi & Connectivity': 'Wifi_Connectivity'}, inplace=True)
df.rename(columns={'Inflight Entertainment': 'Inflight_Entertainment'}, inplace=True)
df.rename(columns={'Value For Money': 'Value_for_Money'}, inplace=True)
```

# Finding the co relation b/w seat comfort and value for money rating given by customers:

```python
seat_comfort_ratings = df['Seat_Comfort']
value_for_money_ratings = df['Value_for_Money']

# Calculating the correlation coefficient
correlation_coefficient = seat_comfort_ratings.corr(value_for_money_ratings)

print("Correlation coefficient between Seat Comfort and Value For Money ratings:", correlation_coefficient)
```

>Insight : Seat comfort and Value of money are having positive corelations.

# Group by analysis to count recommended and non-recommended flights for each seat type

```python
recommended_counts = df[df['Recommended'] == 1].groupby('Seat_Types')['Recommended'].count()
non_recommended_counts = df[df['Recommended'] == 0].groupby('Seat_Types')['Recommended'].count()

# Plotting the stacked bar chart
plt.figure(figsize=(10, 6))
recommended_counts.plot(kind='bar', color='green', label='Recommended')
non_recommended_counts.plot(kind='bar', color='grey', bottom=recommended_counts, label='Non-Recommended')
plt.title('Comparison of Recommended vs. Non-Recommended Flights by Seat Type')
plt.xlabel('Seat Type')
plt.ylabel('Count')
plt.legend()
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group by analysis to calculate percentage of recommended flights for each country
recommended_percentage_by_country = (df[df['Recommended'] == 1].groupby('Country')['Recommended'].count() / df.groupby('Country')['Recommended'].count()) * 100
top_countries_recommended_percentage = recommended_percentage_by_country.nlargest(10)  # Selecting top 10 countries

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
top_countries_recommended_percentage.plot(kind='barh', color='Teal')
plt.title('Top 10 Countries with Highest Percentage of Recommended Flights')
plt.xlabel('Percentage of Recommended Flights')
plt.ylabel('Country')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

 # Corelation matrix and Heatmap

```python
correlation_matrix = df[['Seat_Comfort', 'Ground_Service', 'Value_for_Money', 'Recommended']].corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Set2', fmt=".2f")
plt.title('Correlation Heatmap of Rating Variables')
plt.show()
```

 > Insight : 1. Strong positive correlation between seat comfort and ground service.
 
>           2. Positive correlation between seat comfort and value for money.
