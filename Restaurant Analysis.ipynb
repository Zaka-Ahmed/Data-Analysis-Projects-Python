```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Marketing_Spend', y='Monthly_Revenue', data=df,color='r')
plt.title('Scatter Plot of Marketing Spend vs Monthly Revenue')
plt.show()
```

```python
correlation = df['Marketing_Spend'].corr(df['Monthly_Revenue'])
print(f'Correlation coefficient: {correlation}')
```

```python
#Insight
# Their is weak positive corelation b/w Marketing Spend and Monthly Revenue .Increasing marketing spend will tend to increase monthly revenue, but the effect is not very large.
 
```

```python
# Reviews and Cuisine Type.
# Converting it to Cuisine_Type (Categorical) to Numerical features for more analysis:

mapping = {'Japanese': 0, 'American': 1, 'Mexican': 2, 'Italian':3}
df['Cuisine_Type'] = df['Cuisine_Type'].map(mapping)

df.head()
```

```python
# Bar plot of reviews for each cuisine type
sns.barplot(data=df, x='Cuisine_Type', y='Reviews',palette='dark')
plt.title('Reviews by Cuisine Type')
plt.show()
```

```python
#Insight 
# The most popular cuisine by reviews is  number 3 which is Italian.
```
